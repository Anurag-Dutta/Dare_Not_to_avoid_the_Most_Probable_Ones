```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
csv_file = 'rhine_river_discharge.csv'  # Replace with the actual path to your CSV file
df = pd.read_csv(csv_file)

# Assuming your CSV file has a column named 'Date' and another column named 'Value'
# You can adjust these column names based on your actual data
date_column = 'Year'
value_column = 'Discharge'

# Convert the 'Date' column to datetime format
df[date_column] = pd.DataFrame({'date':pd.date_range(start='1807', end='1956', freq='Y')})

# Sort the DataFrame by date
df.sort_values(by=date_column, inplace=True)

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.size": 18
})

# Plot time series data
plt.figure(figsize=(10, 6))

plt.scatter(df[date_column], df[value_column], color='black', marker='o', label='Markers', s = 10)

plt.plot(df[date_column], df[value_column], color='red', linestyle='-', label='Line')

# plt.plot(df[date_column], df[value_column], marker='o', linestyle='-', color='red', ms = 2)
plt.xlabel('Time Frame', fontsize=18)
plt.ylabel('Discharge (in cubic metres)', fontsize=18)
# plt.title('Time Series Plot')

plt.grid(True)
plt.savefig('rhine_river_discharge.pdf', bbox_inches='tight', labelsize=18)

# Show the plot
plt.grid(True)
plt.show()
```
